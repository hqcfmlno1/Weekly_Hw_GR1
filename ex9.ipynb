```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
n_samples = 100
random_state = 170
center_points = [[1,1], [-1,-1], [1,-1]]
X, y = make_blobs(n_samples = n_samples, random_state = random_state, centers = center_points, cluster_std = 0.6)
# truyền cho tham số centerpoints các điểm dữ liệu, makeblob tự hiểu là cần sinh dữ liệu 2 chiều
```

```python
X
```

```python
X.shape
```

```python
y
```

```python
plt.figure(figsize = (10,10))
plt.scatter(X[:,0], X[:,1], color = 'b')
plt.title("trc khi phân cụm")
plt.show()
```

```python
k_mean_model = KMeans(n_clusters = 3, random_state = random_state)
k_mean_model.fit(X)
```

```python
k_mean_model.cluster_centers_
```

```python
centers = k_mean_model.cluster_centers_
y_pred = k_mean_model.predict(X)
```

```python
y_pred
```

```python
color = ['b', 'g', 'y']
colors = [color[i] for i in y_pred]
plt.figure(figsize = (16,9))
plt.title("sau khi phân cụm")
plt.scatter(X[:,0], X[:,1], color = colors)
plt.scatter(centers[:,0], centers[:,1], color = 'r')
plt.show()
```

```python
# sinh 2 cụm
two_centers = [[1.8,1.8], [3.6,3.6]]
new_X, new_y = make_blobs(n_samples = n_samples, centers = two_centers, random_state = random_state, cluster_std = 0.7)
# std càng cao thì phân độ phân tán của các cụm càng cao
```

```python
plt.figure(figsize = (16,9))
plt.title("100 điểm dữ liệu mới với 2 cụm trước khi phân cụm")
plt.scatter(new_X[:,0], new_X[:,1], color = 'b')
plt.show()
```

```python
# phân cụm với 3 cụm
model_3mean = KMeans(n_clusters = 3, random_state = random_state)
model_3mean.fit(X)
y_pred_3mean = model_3mean.predict(X)
```

```python
# phân cụm với 4 cụm
model_4mean = KMeans(n_clusters = 4, random_state = random_state)
model_4mean.fit(X)
y_pred_4mean = model_4mean.predict(X)
```

```python
# phân cụm với 5 cụm
model_5mean = KMeans(n_clusters = 5, random_state = random_state)
model_5mean.fit(X)
y_pred_5mean = model_5mean.predict(X)
```

```python
color = ['orange', 'b', 'g', 'y', 'purple']
color_3mean = [color[i] for i in y_pred_3mean]
color_4mean = [color[i] for i in y_pred_4mean]
color_5mean = [color[i] for i in y_pred_5mean]
```

```python
# sau khi phân 3 cụm
plt.figure(figsize = (16,9))
plt.title("sau khi phân 3 cụm")
plt.scatter(X[:,0], X[:,1], color = color_3mean)
plt.scatter(model_3mean.cluster_centers_[:,0], model_3mean.cluster_centers_[:,1], color = 'r')
plt.show()
```

```python
# sau khi phân 4 cụm
plt.figure(figsize = (16,9))
plt.title("sau khi phân 4 cụm")
plt.scatter(X[:,0], X[:,1], color = color_4mean)
plt.scatter(model_4mean.cluster_centers_[:,0], model_4mean.cluster_centers_[:,1], color = 'r')
plt.show()
```

```python
# sau khi phân 5 cụm
plt.figure(figsize = (16,9))
plt.title("sau khi phân 5 cụm")
plt.scatter(X[:,0], X[:,1], color = color_5mean)
plt.scatter(model_5mean.cluster_centers_[:,0], model_5mean.cluster_centers_[:,1], color = 'r')
plt.show()
```

# Ứng dụng nén ảnh

```python
import matplotlib.image as image
from IPython.display import Image, display
import os 
```

```python
base_dir = os.getcwd()
chim_be_dir = os.path.join(base_dir, 'Ex9', 'bird_small.png')
display(Image(chim_be_dir, width = 250, unconfined = True))
```

```python
img = image.imread(chim_be_dir)
```

```python
np.max(img) # ảnh png đã được image của matplotlib chuẩn hóa
```

```python
img.shape
```

đang có tất cả 128*128 điểm ảnh và mỗi điểm ảnh có 3 màu
* hiện tại img đang có dạng ma trận 3d, mỗi lớp là ảnh của 1 màu, có tất cả 2 lớp màu
* cần chuyển về ma trận 2d với các cột là 3 điểm màu, các hàng là các điểm ảnh

```python
data_img = img.reshape(-1,img.shape[2]) # ảnh đã được chuẩn hóa không cần chia cho 255 nữa
```

```python
# mỗi tổ hợp các giá trị của rgb tạo thành 1 màu
# kmean tìm ra n màu chủ đạo rồi gán các điểm cùng cụm bằng màu của trung tâm
n_color = 10
k_mean_model = KMeans(n_clusters = n_color, random_state = random_state)
k_mean_model.fit(data_img)
```

```python
print("Số chiều của tâm cụm: ", k_mean_model.cluster_centers_.shape)
print(k_mean_model.cluster_centers_)
print(k_mean_model.labels_[0:20])
# có thể sử dụng label để xem nhãn của các điểm ảnh được sử đụng để train không cần đến predict
```

```python
# đưa các điểm ảnh trong cụm thành màu của trung tâm cụm
image128 = np.array([k_mean_model.cluster_centers_[i] for i in k_mean_model.labels_])
image128 = image128.reshape(128,128,3)
# reshape ảnh về định dạng ban đầu
```

```python
img128_dir = os.path.join(base_dir, 'Ex9', 'image128.png')
old_img_dir = os.path.join(base_dir, 'Ex9', 'bird_small.png')
image.imsave(img128_dir, image128)
```

```python
display(Image(img128_dir, width=250, unconfined=True))
display(Image(old_img_dir, width=250, unconfined=True))
```

```python
print('Size of compressed image: ' + str(os.path.getsize(img128_dir)) + ' KB')
print('Size of original image: ' + str(os.path.getsize(old_img_dir)) + 'KB')
```

# Nén ảnh mới số màu 3
